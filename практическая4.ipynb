```python
class Engine:
    def __init__(self, name, date, country):
        self.name = name
        self.date = date
        self.country = country

    def __str__(self):
        return "\nНазвание двигателя: {}\nДата выпуска: {}\nСтрана производства: {}\nКПД: {}\nТип: {}\nСистема охлаждения: {}".format(self.name, self.date, self.country, self.KPD(), self.type(), self.cooling_systems())

    def KPD(self):
        return '20-90%'

    def type(self):
        return 'Двигатель'

    def cooling_systems(self):
        return True


class DVS(Engine):
    def __init__(self, name, date, country, fuel):
        super().__init__(name, date, country)
        self.fuel = fuel

    def KPD(self):
        return '20-25%'

    def type(self):
        return 'Двигатель внутреннего сгорания'


class RD(Engine):
    def __init__(self, name, date, country, G, C, U):
        super().__init__(name, date, country)
        self.G = G  # Секундный расход массы рабочего тела через двигатель
        self.C = C  # Скорость
        self.U = U  # секундный расход массы рабочего тела через двигатель

    def P(self):
        P = self.G * (self.C - self.U)
        return "Реактивная тяга = {}(Н)".format(P)

    def KPD(self):
        return '47-60%'

    def type(self):
        return 'Реактивный двигатель'


class Diesel(DVS):
    def __init__(self, name, date, country, fuel, litr):
        super().__init__(name, date, country, fuel)
        self.litr = litr

    def KPD(self):
        return '40-45%'

    def type(self):
        return 'Дизель'

m1 = Engine('motor_1', 1807, 'France')
print(m1)
m2 = DVS('motor_2', 1900, 'USSR', 'АИ-95')
print(m2, m2.fuel, sep="\n")
m3 = RD('motor_3', 2004, 'Russia', 200, 760000, 230000)
print(m3, m3.P(), sep="\n")
m4 = Diesel('motor_4', 2002, 'Russia', 'Аи-92', 'V-8')
print(m4, m4.fuel, m4.litr, sep="\n")
# obj1.init

# for Engine in (m1, m2, m3, m4):
# print(Engine)
```
