Выведите среднюю температуру за наблюдаемый промежуток времени, предварительно очистив список от неопределенных значений. Гарантируется, что хотя бы одно определенное значение в списке есть.

```python
# Программирование на языке высокого уровня (Python).
# https://www.yuripetrov.ru/edu/python
# Задание task_05_02_02.
#
# Выполнил: Сачкова Г.Г
# Группа: АДЭУ211


def avg(data):

    data = sum(data) / len(data)
    return data


def cleared_data(data):

    data.remove(None)
    return data


n = int(input("Кол-во измерений: "))
# Если с клавиатуры вводится прочерк, измерения нет
# (необходимо добавить в список None)
data = []
for i in range(1, n + 1):
    x = input("Измерение {}-е: ".format(i))
    if x == '-' or x == '':
        data.append(None)
    else:
        data.append(float(x))
# Получить очищенный список и среднее значение
print("Средняя температура: " + "{0:.2f}".format(avg(cleared_data(data))))
# --------------
# Пример вывода:
#
# Кол-во измерений: 3
# Измерение 1-е: 10
# Измерение 2-е: -
# Измерение 3-е: 20
# Средняя температура: 15.00
```

№ 5.2.11

Используя шифр Цезаря (достаточно только букв русского алфавита, знаки препинания не изменяются), зашифруйте, а затем расшифруйте введенную строку.

```python
# Программирование на языке высокого уровня (Python).
# https://www.yuripetrov.ru/edu/python
# Задание task_05_02_11.
#
# Выполнил: Сачкова Г.Г.
# Группа: АДЭУ211
# E-mail: !!!


def ceasar(text, shift):
    """Вернуть измененную строку 'text' со сдвигом 'shift'.

    Параметры:
        - text (str): строка;
        - shift (int): свдиг.

    Результат:
        str: измененная строка."""
    # Набор кириллических букв
    result = []
    letters = [chr(i) for i in range(ord('А'), ord('я') + 1)]
    for i in text:
        if i in letters:
            # print(ord(i), i)
            i = ord(i) + shift
            if i > 1103:
                i -= 64
            if i < 1040:
                i += 64
            i = chr(i)
            result.append(i)
            # print(i)
        else:
            result.append(i)
    result = ''.join(result)
    return result


text = str(input('Введите предложение: '))
shift = int(input('Введите сдвиг: '))
ceasar(text, shift)
encoded = ceasar(text, shift)
decoded = ceasar(encoded, shift * -1)
print("Зашифрованная строка:", encoded)
print("Расшифрованная строка:", decoded)

# --------------
# Пример вывода:
#
# Введите предложение: ПрограММиРОВание С++
# Введите сдвиг: 4
# Зашифрованная строка: УфтзфдРРмФТЖдсмй Х++
# Расшифрованная строка: ПрограММиРОВание С++
```

отрицательных значений (отсортирован по убыванию);
неотрицательных значений (отсортирован по возрастанию).


```python
# Программирование на языке высокого уровня (Python).
# https://www.yuripetrov.ru/edu/python
# Задание task_05_02_14.
#
# Выполнил: Сачкова Г.Г.
# Группа: АДЭУ211
# E-mail: !!!


def split_numbers(*nums):
    """Вернуть 2 списка - отрицательных и неотицательных значений.

    !!!

    Результат:
        - list: отрицательные значения;
        - list: неотицательные значения."""
    plus = tuple()  # Положительные
    minus = tuple()  # Отрицательные
    for i in nums:
        if i >= 0:
            plus += (i,)
        else:
            minus += (i,)
    minus = sorted(minus, reverse=True)
    plus = sorted(plus)
    result = (minus, plus)
    return result


print(split_numbers(1, 4, -5, 0, -33))
# --------------
# Пример вывода:
#
# ([-5, -33], [0, 1, 4])
```

```python
# Программирование на языке высокого уровня (Python).
# https://www.yuripetrov.ru/edu/python
# Задание task_05_02_16.
#
# Выполнил: Фамилия И.О.
# Группа: !!!
# E-mail: !!!


def pow1(value, power):
    """Вернуть 'value' в степени 'power'.

    Параметры:
        - !!!.

    Результат:
        !!!."""
    
    v=value
    p=power
    return v**p #"""Вернуть 'value' в степени 'power'


def pow2(value, power):
    """Вернуть 'value' в степени 'power'. Рекурсивный алгоритм.

    Параметры:
        - !!!.

    Результат:
        !!!."""

    v=value
    p=power
    if v==0 or v==1:
      return v
    else:
      return (v**p)*pow2(1,1)


print(pow1(5, 3))
print(pow2(5, 3))

# --------------
# Пример вывода:
#
# 125
# 125
```

№ 5.2.17

Дано натуральное число. Напишите рекурсивные функции для определения:

суммы цифр числа;
количества цифр в числе.


```python
# Программирование на языке высокого уровня (Python).
# https://www.yuripetrov.ru/edu/python
# Задание task_05_02_17.
#
# Выполнил: Сачкова Г.Г.
# Группа: АДЭУ211
# E-mail: !!!


def digits_sum(value):
  v = value
  """Вернуть сумму цифр числа 'value'. Рекурсивная реализация.

    Параметры:
        - !!!.

    Результат:
        !!!.
    """
  if v > 0:
      return digits_sum(v // 10) + v % 10
  else:
      return 0


def digits_count(value):
  v = value
  """Вернуть количество цифр числа 'value'. Рекурсивная реализация.

    Параметры:
        - !!!.

    Результат:
        !!!.
    """
  if v > 0:
    return digits_count(v // 10) + 1
  else:
    return 0


print(digits_sum(12345))
print(digits_count(12345))

# --------------
# Пример вывода:
#
# 15
# 5
```
