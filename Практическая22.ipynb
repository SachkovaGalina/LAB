```python
class Engine:
    def __init__(self, name, date, country):
        self.name = name
        self.date = date
        self.country = country
    def __str__(self):
        return "Название двигателя: {}\nДата выпуска: {}\nСтрана производства: {}\nКПД: {}\nТип: {}\nСистема " \
               "охлаждения: {}\n".\
            format(self.name, self.date, self.country, self.KPD(), self.type(), self.cooling_systems())
    def KPD(self):
        return '20-90%'
    def type(self):
        return 'Двигатель'
    def cooling_systems(self):
        return True
class DVS(Engine):
    def __init__(self, name, date, country):
        super().__init__(name, date, country)
    def KPD(self):
        return '20-25%'
    def type(self):
        return 'Двигатель внутреннего сгорания'
class RD(Engine):
    def __init__(self, name, date, country):
        super().__init__(name, date, country)
    def KPD(self):
        return '47-60%'
    def type(self):
        return 'Реактивный двигатель'
class Diesel(DVS):
    def __init__(self, name, date, country):
        super().__init__(name, date, country)
    def KPD(self):
        return '40-45%'
    def type(self):
        return 'Дизель'
m1 = Engine('motor_1', 1807, 'France')
print(m1)
m2 = DVS('motor_2', 1900, 'USSR')
print(m2)
m3 = RD('motor_3', 2004, 'Russia')
print(m3)
m4 = Diesel('motor_4', 2002, 'Russia')
print(m4)
```
